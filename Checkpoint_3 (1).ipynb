```python
#ViewCorrelation
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
#Prepare the features and target
#Get the features by returning the dataframe without the target column
x=data.drop("price", axis=1)

#Get the target by directly selecting the column
y=data["price"]

#Dealing with dates:
#First convert the date into a python date format 
x["date"] = pd.to_datetime(x["date"])

#Then separate date/time into distinct features
x["year"] = x["date"].dt.year
x["month"] = x["date"].dt.month
x["week"] = x["date"].dt.week
x["day"] = x["date"].dt.day
x["hour"] = x["date"].dt.hour
x["minute"] = x["date"].dt.minute
x["day_of_week"] = x["date"].dt.dayofweek

#And finally delete the date column
x = x.drop("date", axis=1)
x = x.drop("zipcode", axis=1)
x = x.drop("yr_built", axis=1)
x = x.drop("yr_renovated", axis=1)


x_train,x_test,y_train,y_test,=train_test_split(x,y, test_size=0.20, random_state=30)

print(x)
```

```python
#4. Apply Linear regression to your training set. 
model=LinearRegression()  
model.fit(x_train,y_train)
```

```python
#Plot the linear regression. 
import matplotlib.pyplot as plt
plt.scatter("bedrooms","price",color="r")
plt.title("Linear Regression")
plt.ylabel("Price")
plt.xlabel("bedrooms")
plt.plot(x,model.predict(x),color="b")
plt.show()
```

```python
#Measure the performance of linear regression using the testing set. 
predicted=model.predict(x_test)
print(predicted)
```

```python
#Apply multi-linear regression and compare it to the linear model. 
#extract x and y from our data
x=data[["bedrooms","bathrooms","sqft_living","sqft_lot","floors","waterfront","view","condition","grade","sqft_above","sqft_basement","yr_built","yr_renovated","zipcode","lat","long"	,"sqft_living15", "sqft_lot15"]]  #we have more than one input
y=data["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) #splitting data with test size of 35%

model=LinearRegression() #build linear regression model
model.fit(x_train,y_train) #fitting the training data
predicted=model.predict(x_test) #testing our model’s performance
print("MSE =", mean_squared_error(y_test,predicted))
print("R squared =", metrics.r2_score(y_test,predicted))
```
